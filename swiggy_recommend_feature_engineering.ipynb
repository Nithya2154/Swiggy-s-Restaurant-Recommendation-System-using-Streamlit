### Step 2: Feature Engineering

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('cleaned_data.csv')
df.head()
```

####  2.1 Split Cuisine

```python
df['cuisine'] = df['cuisine'].str.split(',')
```

```python
df.head()
```

#### 2.2 Encode Cuisine

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
cuisine_encoded = mlb.fit_transform(df['cuisine'])
cuisine_df = pd.DataFrame(cuisine_encoded, columns=mlb.classes_)
```

#### 2.3 Encode City

```python
city_encoded = pd.get_dummies(df['city'], prefix='city')
maincity_encoded = pd.get_dummies(df['maincity'], prefix='maincity')
```

#### 2.4 Combine Features

```python
final_df = pd.concat([df[['cost', 'rating']],   # numeric features
    city_encoded,
    cuisine_df
], axis=1)
```

####  2.5 Scale Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
final_df[['cost', 'rating']] = scaler.fit_transform(final_df[['cost', 'rating']])
```

#### 2.6 Final Check

```python
print(final_df.isnull().sum().sum())   # should be 0
print(final_df.dtypes)                 # all numeric
```

#### 2.7 Save Encoded Data

```python
final_df.to_csv('encoded_data.csv', index=False)
```
